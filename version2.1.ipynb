```python
import pandas as pd
import pickle
from sklearn.metrics.pairwise import cosine_similarity
```

```python
with open('data/interim/books_data.pkl', 'rb') as f:
    books_data = pickle.load(f)

with open('data/interim/ratings_data.pkl', 'rb') as f:
    ratings_data= pickle.load(f)

with open('data/interim/users_data.pkl', 'rb') as f:
    users_data= pickle.load(f)
```

```python
# create sample data
sample = ratings_data.sample(n=1000, random_state=42)
ISBN = sample.iloc[1]['ISBN']
```

```python
ratings_data = ratings_data[ratings_data['ISBN'].isin(['0590353403', '0439136350', '0439139597','0439064864', '8445071408'])]
ratings_data
```

```python
concatenated_df = pd.concat([ratings_data, sample], ignore_index=True)

concatenated_df = concatenated_df.drop_duplicates()

concatenated_df
```

```python
pivoted_df = concatenated_df.pivot(index='ISBN', columns='User-ID', values='Book-Rating')
pivoted_df.fillna(0, inplace=True)
pivoted_df

## pivoted_df.loc[['8445071408']][92]
```

```python
cosine = pd.DataFrame(cosine_similarity(pivoted_df)) 
```
