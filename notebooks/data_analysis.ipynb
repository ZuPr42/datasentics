```python
books_ratings_merge = pd.merge(books_data, ratings_data, on='ISBN', how='left')
books_ratings_merge
```

```python
mean_rating = books_ratings_merge.groupby('ISBN')['Book-Rating'].mean().reset_index(name = 'MeanRating')

books_ratings = pd.merge(books_data, mean_rating, how= 'inner', on = 'ISBN')

books_ratings
```

```python
number_of_ratings= books_ratings_merge.groupby('ISBN')['Book-Rating'].count().reset_index(name = 'NumberOfRatings')

books_ratings = pd.merge(books_ratings, number_of_ratings, how= 'inner', on = 'ISBN')

books_ratings
```

```python
books_ratings.sort_values(by=['NumberOfRatings','MeanRating'], ascending=[False, False])
```
