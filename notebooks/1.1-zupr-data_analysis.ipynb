```python
import pandas as pd
import pickle
```

```python
# import of the data
with open('C:/Users/ASUS/anaconda3/data_scentics/data/interim/books_data.pkl', 'rb') as f:
    books_data = pickle.load(f)

with open('C:/Users/ASUS/anaconda3/data_scentics/data/interim/ratings_data.pkl', 'rb') as f:
    ratings_data= pickle.load(f)

with open('C:/Users/ASUS/anaconda3/data_scentics/data/interim/users_data.pkl', 'rb') as f:
    users_data= pickle.load(f)
```

```python
# left merge/join of ratings data and books data
pd.merge(ratings_data, books_data, on='ISBN', how='left')
```

```python
# left merge/join of books data and ratings data
books_ratings_merge = pd.merge(books_data, ratings_data, on='ISBN', how='left')
books_ratings_merge
```

```python
# adding column with mean rating to books data
mean_rating = books_ratings_merge.groupby('ISBN')['Book-Rating'].mean().reset_index(name = 'MeanRating')

books_ratings = pd.merge(books_data, mean_rating, how= 'left', on = 'ISBN')

books_ratings
```

```python
# adding column with number of ratings to books data
number_of_ratings= books_ratings_merge.groupby('ISBN')['Book-Rating'].count().reset_index(name = 'NumberOfRatings')

books_ratings = pd.merge(books_ratings, number_of_ratings, how= 'left', on = 'ISBN')

books_ratings
```

```python
# saving the merged data
with open('C:/Users/ASUS/anaconda3/data_scentics/data/interim/books_ratings.pkl', 'wb') as f:
    pickle.dump(books_ratings, f)
```
