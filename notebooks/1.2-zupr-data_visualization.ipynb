```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
# load data
with open('C:/Users/ASUS/anaconda3/data_scentics/data/interim/books_data.pkl', 'rb') as f:
    books_data = pickle.load(f)

with open('C:/Users/ASUS/anaconda3/data_scentics/data/interim/ratings_data.pkl', 'rb') as f:
    ratings_data= pickle.load(f)

with open('C:/Users/ASUS/anaconda3/data_scentics/data/interim/users_data.pkl', 'rb') as f:
    users_data= pickle.load(f)

with open('C:/Users/ASUS/anaconda3/data_scentics/data/interim/books_ratings.pkl', 'rb') as f:
    books_ratings= pickle.load(f)
```

```python
books_ratings
```

```python
# select data with the number of books per year
data = books_ratings.groupby('Year-Of-Publication')['ISBN'].count().reset_index(name = 'Count').sort_values(by = ['Year-Of-Publication'])

data
```

```python
# Plot data with number of books
data.plot(x='Year-Of-Publication', y='Count', kind='line')
plt.xlabel('Year-Of-Publication')
plt.ylabel('Count')
plt.title('Data Visualization')
plt.show()
```

```python
# select data with the last 20 years of rating
data = books_ratings.groupby('Year-Of-Publication')['MeanRating'].mean().reset_index(name = 'Rating').sort_values(by = ['Year-Of-Publication'], ascending = False).head(20)

data
```

```python
# Plot data of ratings
data.plot(x='Year-Of-Publication', y='Rating', kind='bar')
plt.xlabel('Year-Of-Publication')
plt.ylabel('Rating')
plt.title('Data Visualization')
plt.show()
```
