```python
import pandas as pd
import pickle
```

```python
with open('C:/Users/ASUS/anaconda3/data_scentics/data/interim/books_data.pkl', 'rb') as f:
    books_data = pickle.load(f)

with open('C:/Users/ASUS/anaconda3/data_scentics/data/interim/ratings_data.pkl', 'rb') as f:
    ratings_data= pickle.load(f)

with open('C:/Users/ASUS/anaconda3/data_scentics/data/interim/users_data.pkl', 'rb') as f:
    users_data= pickle.load(f)

```

```python
# create sample data
sample = ratings_data.sample(n=1000, random_state=42)
ISBN = sample.iloc[1]['ISBN']
```

```python
def correlation(ISBN):

    # Filtered subset of ratings_data based on ISBN
    ratings_data_filtered = ratings_data[ratings_data['ISBN'].isin(['0590353403', '0439136350', '0439139597','0439064864', '8445071408'])]

    # Concatenate the filtered subset with the sample DataFrame
    concatenated_df = pd.concat([ratings_data_filtered, sample], ignore_index=True)
    concatenated_df = concatenated_df.drop_duplicates()

    # create pivoted matrix
    pivoted_df = concatenated_df.pivot(index='User-ID', columns='ISBN', values='Book-Rating')
    pivoted_df.fillna(0, inplace=True)
    #pivoted_df

    # calculate correlation matrix
    correlation = pivoted_df.corr()

    correlation = correlation.loc[ISBN].reset_index(name = 'Index')

    correlation  = correlation[(correlation['Index'] > 0) & (correlation['Index'] < 1)]

    books = books_data.loc[books_data['ISBN'].isin(correlation['ISBN'])]

    return(books)
```
